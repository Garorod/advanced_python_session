# Twitter API

```python
import re

try:
    import json
except ImportError:
    import simplejson as json

import twitter
import urllib 
import pprint
import time
import yaml
```

```python
# !pip install python-twitter
```

## import credentials

```python
with open('twitter.yaml', 'r') as f:
    key = yaml.load(f, yaml.BaseLoader)
```

## Initialize twitter API with credentials 

```python
t = twitter.Api(consumer_key=key['API key'],
                consumer_secret=key['API secret key'],
                access_token_key=key['Access Token'],
                access_token_secret=key['Access Token Secret'])
```

## Save the data into a file

```python
ts = int(time.time())
stocks = ['#GOOG', '#AMZN', '#AAPL', '#TSLA', '#FB', '#MSFT', '#IBM', '#SNAP', '#BOX', '#NFLX']
with open(str(ts) + "-tweets.txt", 'w') as fo:
    for stock in stocks:
        results = t.GetSearch(raw_query="q="+urllib.parse.quote(stock), count=5, lang='eng')
        for tweet in results:
            fo.write(json.dumps(json.loads(tweet.AsJsonString())) + '\n')
```

### Open the file to read the data 

```python
with open(str(ts) + "-tweets.txt", 'rb') as tweets_file:
    for line in tweets_file:
        try:
        # Read in one line of the file, convert it into a json object
            tweet = json.loads(line.strip())
            if 'text' in tweet: # only messages contains 'text' field is a tweet
                print (tweet['id']) # This is the tweet's id
                print (tweet['created_at']) # when the tweet posted
                re.sub(' +', ' ', tweet['text'])
                
                print (tweet['text']) # content of the tweet
                hashtags = []
                for hashtag in tweet['entities']['hashtags']:
                    hashtags.append(hashtag['text'])
                print (hashtags)
                print ("\n")

        except:
            # read in a line is not in JSON format (sometimes error occured)
            continue
```
